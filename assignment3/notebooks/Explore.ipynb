```python
import pandas as pd
import numpy as np
```

```python
reservoir = np.array([])
tags = np.array([])
k = 1000

def store_smallest(reservoir, tags, new_ip, new_tag, k):
    reservoir = np.append(reservoir, new_ip)
    tags = np.append(tags, new_tag)
        
    ind = np.argsort(tags)[:k]
    return (reservoir[ind], tags[ind])

with open("../data/capture20110815-2.pcap.netflow_46_.labeled", "r") as ins:
    i = 0
    for line in ins:
        parts = line.split()
        ip_port_src = parts[4].split(':')
        ip_src = ip_port_src[0]
        r = np.random.rand()
        
        (reservoir, tags) = store_smallest(reservoir, tags, ip_src, r, k)
```

```python
a, b = np.unique(reservoir, return_counts=True)
print(a)
print(b)
```

## Figure out true top 10

```python
ips = {}
with open("../data/capture20110815-2.pcap.netflow_46_.labeled", "r") as ins:
    for line in ins:
        parts = line.split()
        ip_port_src = parts[4].split(':')
        ip_src = ip_port_src[0]
        if not ip_src in ips:
            ips[ip_src] = 0
        ips[ip_src] += 1
```

```python
k = np.array(list(ips.keys()))
v = np.array(list(ips.values()))

x = np.argsort(v)
s = k[x]
s_v = v[x]
print(s[len(s)-10:])
print(s_v[len(s)-10:])
```
